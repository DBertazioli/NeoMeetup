### Set the URL you want to webscrape from

```python
url = 'https://www.meetup.com/topics/'
```

### Connect to the URL

```python
response = requests.get(url)
```

### Parse HTML and save to BeautifulSoup object

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
category_list=[]
for i in range(len(soup.findAll('a'))): 
    one_a_tag = soup.findAll('a')[i]
    try:
        link = one_a_tag['name']
        #print(i)
        category_list.append(link)
    except Exception as e:
        pass
    
    #time.sleep(1)
    
```

```python
category_list
```

```python
topic_list=[]
it=0
topic_dict={}
for i in range(6,len(soup.findAll('a'))): #be careful, working starting with 7 cause of reasons
    one_a_tag = soup.findAll('a')[i]
    try:
        name=one_a_tag['name']
    except Exception:
        pass
    try:
        link = one_a_tag['href']
        link = link.replace("/topics/","")
        link = link.replace("/","")
        topic_list.append(link)#checking list
        topic_dict[link]=name
        if "nodejs" in link:
            print("reached the end")
            break
    except Exception as e:
        pass
        
    
```
